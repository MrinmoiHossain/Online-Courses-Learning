```python
import numpy as np
import pandas as pd
```

```python
df_crime = pd.read_csv('Police_Department_Incidents_-_Previous_Year__2016_.csv')
df_crime.head()
```

```python
df_crime.drop(['IncidntNum', 'Category', 'Descript', 'DayOfWeek', 'Date', 'Time', 'Resolution', 'Address', 'X', 'Y', 'Location', 'PdId'], axis = 1, inplace = True)
df_crime.rename(columns={'PdDistrict':'Neighborhoods'},  inplace = True)
df_crime.head()
```

```python
df_crime_added = df_crime['Neighborhoods'].value_counts()
df_crime_added
```

```python
df_crime_list = pd.DataFrame({'Neighborhoods': ['SOUTHERN','NORTHERN','MISSION','CENTRAL','BAYVIEW','INGLESIDE','TARAVAL',
                                          'TENDERLOIN','RICHMOND','PARK'],
                   'Count':  [28445, 20100, 19503, 17666, 14303, 11594, 11325, 9942, 8922, 8699]})
df_crime_list.sort_index(axis = 1, level = 'Neighborhoods', inplace = True, ascending = False)
df_crime_list
```

```python
import folium
```

```python
world_geo = r'san-francisco.geojson'

threshold_scale = np.linspace(df_crime_list['Count'].min(), df_crime_list['Count'].max(), 6, dtype=int)

world_map = folium.Map(location=[37.77,-122.42], zoom_start = 12)

world_map.choropleth(
    geo_data = world_geo,
    data = df_crime_list,
    columns = ['Neighborhoods', 'Count'],
    key_on = 'feature.properties.DISTRICT',
    fill_color = 'YlOrRd', 
    fill_opacity  =0.7, 
    line_opacity = 0.2,
    legend_name = 'Crime Rate in San Francisco',
    reset = True
)
world_map
```
